```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("MuRun2010B_withoutM.csv", sep="\s+")
df["invariant_mass"] = np.sqrt(
    (df["E1"] + df["E2"])**2
    - (df["px1"] + df["px2"])**2
    - (df["py1"] + df["py2"])**2
    - (df["pz1"] + df["pz2"])**2
)

plt.figure(figsize=(10,6))
plt.hist(df["invariant_mass"], bins=150, range=(0,150), alpha=0.7, color='blue')
plt.title("CMS Data: Dimuon Invariant Mass Spectrum")
plt.xlabel("Invariant Mass (GeV)")
plt.ylabel("Counts")
plt.axvline(3.1, color='red', linestyle='dashed', label="J/ψ (3.1 GeV)")
plt.axvline(9.5, color='green', linestyle='dashed', label="Υ(1S) (9.5 GeV)")
plt.axvline(91, color='purple', linestyle='dashed', label="Z boson (91 GeV)")
plt.yscale("log")

plt.legend()
plt.grid(True)
plt.show()
```
